```python
import ibis
import ibis.selectors as s
from ibis import _
ibis.options.interactive = True # this line show the first few rows for saving computatino
starwars = ibis.examples.starwars.fetch()
```

```python
(
    starwars
        .filter(_.skin_color=='light') #_ is short hand for table from previous steps
)
```

```python
(starwars
    .mutate(
        height_m = _.height / 100
    )
    .mutate(
        BMI = _.mass / (_.height_m**2)
    )
    .select("BMI", ~s.matches("BMI")) #don't be fooled, '~' is just negative in python
)
```
