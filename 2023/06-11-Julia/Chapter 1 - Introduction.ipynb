### Introduction

Keen feature of Julia: 

- list apprehension (create a list on the fly with a loop)
- define a function with `x -> f` 
    - remember in formal definition function close with end
    - `map` `broadcast` first argument is always a function 
    - put '.' dot after function similar has samilar effect as .

```julia
v = [i for i in 1:10]
power_2 = x -> x^2
power_2(2)
power_2.(v)

using Plots
using PlotlyJS
# plotlyjs()
plot
plot(v, power_2.(v))
```
